# Configurando Ambiente

```python
import basedosdados as bd
```

```python
# Salvando a tabela em uma variavel do projeto
df = bd.read_table(dataset_id='br_abrinq_oca',
                    table_id='municipio_primeira_infancia',
                    billing_project_id="api-ads-388916",
                    reauth=True)
```

# Gerando a Busca

```python
# Solicitando os dados para o usuario
municipio = input("Informe o código do municipio: ")
baixar_tabela = input("Informe apenas com sim ou nao (conforme está escrito) parabaixar ou não a pesquisa : ")

# Filtrando o municipio pelo código
tabela_busca = df[df['id_municipio'] == municipio]

# Baixando a tabela
if baixar_tabela == 'sim' or baixar_tabela == 'Sim':
    tabela_busca.to_excel(f'criancas_nas_escolas_{ano}_{municipio}.xlsx')
elif baixar_tabela == 'nao' or baixar_tabela == 'Nao':
    display(tabela_busca)
else:
    print("Tente novamente, favor responder conforme orientado!")
```
